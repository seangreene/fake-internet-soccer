```python
import pandas as pd
import altair as alt
```

```python
df12 = pd.read_csv("data/Fantrax-Transaction-History-Trades-Fake Internet Soccer XII.csv")
```

```python
df12.head()
```

```python
df12.groupby(["From","To"])["Player"].size().reset_index().sort_values("Player", ascending=False).rename(columns={"Player":"Trades"})
```

```python
alt.Chart(
    df12.groupby(["From","To"])["Player"].size().reset_index()    
).mark_point().encode(
    x="From",
    y="To",
    size="Player"
).properties(width=400, height=400)
```
